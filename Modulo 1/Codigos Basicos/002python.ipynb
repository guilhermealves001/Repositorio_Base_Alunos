#função print ()
A funçao print () é uma das mais importantes eutilizadas na linguagem python. Sua funçao e basicamente exibir mensagens na tela ou envia-las para outros dispositivos, como imprimir dentro de um arquivo de texto.
No python 3, print() é uma funçaõ interna, de modo que não é necessario importar de nenhuma outra biblioteca para poder utiliza-la, basta chamar ela e passar os argumentos necessarios

```python
print("meu primeiro print") #todos os textos devem estar em aspas simples ou duplas
```

```python
print("meu segundo print") #
```

```python
print("D`agua") #quando uma palavra tem aspas, as aspas iniciais devem ser diferentes
```

```python
print(20+15) #o python realiza a operação que esta no parenteses
```

```python
print(3567+84364)
```

```python
print(23567-8252)
```

```python
print("20+15") #a operação entre aspas sera apenas impressa, mas não executada, pois o python entende isso como texto
```

```python
print("20+15", 20+15)
```

```python
n1 = "1"
n2 = "3"
print(n1+n2) #neste exemplo o python faz a concatenação das duas strings
```

```python
n1 = 1
n2 = 3
print(n1+n2)
```

```python
nome = "ronaldinho"
sobrenome = "gaúcho"
nome_completo = nome + " " + sobrenome
print(nome_completo)
```
