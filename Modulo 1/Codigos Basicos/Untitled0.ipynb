```python
temper =int(input("Digite a temperatura do dia:"))
if temper >= 30:
    print(f"A temperatura do dia é {temper}°C e o dia está muito quente")
elif temper >= 20:
    print(f"A temperatura do dia é {temper}°C e o dia está agradável")
elif temper >= 10:
    print(f"A temperatura do dia é {temper}°C e o dia está frio")
```
