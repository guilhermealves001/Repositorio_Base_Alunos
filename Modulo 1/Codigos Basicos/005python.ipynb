## Condicional simples
* Para condições em python utilizamos a palavra "if" que significa se em português
* Com a condicional, você consegue executar um conjunto de código apenas se o resultado for verdadeiro (true)

```python
idade = int(input("Qual a sua idade? "))
if idade >= 18:
    print("Você é maior de idade")
else:
    print("Você é menor de idade")
```

Crie um código que receba 3 notas, calcule a média informe se o aluno está aprovado ou reprovado
* Obs:Média 7 para aprovar
* Média 4 recuperação média
* Etapas: solicitar as notas do usuário, calcular a média, checar a condição do aluno, informar o resultado

```python
n1 = float(input("Digite a primeira nota: "))
n2 = float(input("Digite a segunda nota: "))
n3 = float(input("Digite a terceira nota: "))
media = (n1 + n2 + n3) / 3
print("A média do aluno é: ", media)
if media >= 7:
    print("O aluno está aprovado")
elif media >= 4:
    print("O aluno está em recuperação")
else:
    print("O aluno está reprovado")
```
