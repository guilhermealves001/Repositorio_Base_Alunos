4) Crie uma calculadora simples (adição, subtração, multiplicação e divisão). Seu código deve receber 2 números e apresentar o resultado das 4 operações

```python
num1 = float(input("Digite um número: "))
num2 = float(input("digite um outro número:" ))
soma = num1 + num2
subtracao = num1 - num2
multiplicacao = num1 * num2
divisao = num1 / num2
print(f"O resultado da soma é {soma}, da subtração é {subtracao}, da multiplicação é {multiplicacao} e da divisão é {divisao}.")
```
